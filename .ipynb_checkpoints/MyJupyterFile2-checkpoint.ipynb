```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
dataset = pd.read_csv('house_data.csv')
dataset.head()
```

```python


X = dataset.iloc[: , 5:6].values


y = dataset.iloc[: , 2:3].values

m = len(y)

X = np.concatenate((np.ones((m,1), dtype=int), X), axis=1) 
print(X.shape)
print(X)
print(y)
```

```python
theta = np.array([[0.0 for i in range(len(X.columns))]]).T
x = X
alpha = 0.01
m = len(X)
```

```python
for i in range(20000):
    y_predict = x@theta
    e = (y_predict - y)
    error = np.mean(abs(e))
    
    theta = theta - alpha/m * x.T@(y_predict-y)
    if i%100 == 0:
        print("theta", theta)
        print("error", error)
```
